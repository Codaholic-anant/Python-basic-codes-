**1**. **SWAP OF 2 INTEGERS**

```python
num1 = int(input("enter the first number"));
num2 = int(input("enter the second number"));
#for swap the number
num = num1;
num1 = num2;
num2 = num;

print(f"num1={num1} , num2={num2}")
```
