**8. check the number is a prime or not**

```python
num = int(input("enter a number"));
is_prime = False
if num<=1:
  print(num,"is not a prime");

elif num>1:
  for i in range(2,num):
    if num % i == 0:
      is_prime = True
      break

if is_prime:
  print(num,"is not a prime number");
else:
  print(num,"is a prime number");

```

**9. check the series is a prime or not**

```python
num = int(input("enter a number you wanna print series"));

is_prime = [];

if num<=1:
  print(num,"is not a prime");

elif num>1:
  for i in range(2 , num+1):
    for j in range(2,i):
      if i % j == 0:
        break
    else:
      is_prime.append(i)

print(is_prime)


```
