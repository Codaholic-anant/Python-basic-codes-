**@to reverse a string**

```python
class String:
  def __init__(self,string):
    self.string = string
    self.reverse = ""

  def __str__(self):
    return f"string is :{self.string} , reverse of the string is :{self.reverse}"

  def reverse_string(self):

    try:
      if self.string == " " or self.string == type(int):
        raise ValueError("Error")
      else:
        for i in range(len(self.string)-1,-1,-1):
          self.reverse += self.string[i]
    except Exception as e:
      print(e)

if __name__ == "__main__":
  my_string = String("hello")
  my_string.reverse_string()
  print(my_string)


```

**@ to check for palindrome**

```python
class Palindrome:
  def __init__(self,text):
    self.text = text
    self.text1 = ""
  def __str__(self):
      return f"text is palindrome {self.text}||reverse is :{self.text1}"

  def check_palindrome(self):

    try:

      if self.text == " " or self.text == type(int):
        raise ValueError("Error")
      else:


        for i in range(len(self.text)-1,-1,-1):
          self.text1 += self.text[i]
        if self.text.lower() == self.text1.lower():
            print(f"{self.text} is a palindrome")
            # return f"{self.text} is a palindrome"
        else:
            print(f"{self.text} is not a palindrome")
            # return f"{self.text} is a not palindrome"
    except Exception as e:
      print(e)







if __name__ == "__main__":
  my_text = Palindrome("Radar")
  my_text.check_palindrome()
  print(my_text)




```

**@Fibonaci series**

```python
class Fibonaci:
  def __init__(self,num):
    self.num = num
    self.series = []

  def check_fibonaci(self):
    try:
      for i in range(self.num):
        if i == 0 or i ==1:
          self.series.append(i)
        else:
          self.series.append(self.series[i-1]+self.series[i-2])

      print(self.series)
    except Exception as e:
      print(e)

if __name__ == "__main__":
  my_num = Fibonaci(50)
  my_num.check_fibonaci()
```

**@Factorial number**

```python
class Factorial:
    def __init__(self, num):
        self.num = num
        self.factorial = 1

    def find_factorial(self):
        try:
            # Negative numbers don't have factorials.
            if self.num < 0:
                raise ValueError("Factorial does not exist for negative numbers")

            elif self.num == 0 or self.num == 1:
                self.factorial = 1

            else:

                for i in range(1, self.num + 1):
                    self.factorial *= i  # Multiply and assign

            print(f"The factorial of {self.num} is {self.factorial}")

        except Exception as e:
            print(f"Error: {e}")

if __name__ == "__main__":
    my_num = Factorial(5) # 5! = 120
    my_num.find_factorial()
```

**@ Print max and min numbers in a list**

```python
class ListNumber:
    def __init__(self, list_numbers):
        self.list_numbers = list_numbers
        self.max_val = None
        self.min_val = None

    def find_max_min(self):
        try:
            # Check if list is empty
            if not self.list_numbers:
                return "List is empty"

            # 1. Assume the first number is both Max and Min
            self.max_val = self.list_numbers[0]
            self.min_val = self.list_numbers[0]

            # 2. Loop through the list once
            for num in self.list_numbers:
                if num > self.max_val:
                    self.max_val = num
                if num < self.min_val:
                    self.min_val = num

            return f"Max is {self.max_val} and Min is {self.min_val}"

        except Exception as e:
            return f"Error: {e}"

if __name__ == "__main__":
    my_list = ListNumber([10, 2, 33, 4, 5])
    result = my_list.find_max_min()
    print(result)
```

**@ remove duplicate in a list**

```python
class Duplicate:
  def __init__(self,data):
    self.data = data
    self.arrow = []
  def remove_duplicate(self):
    try:
      for val in self.data:
        if val not in self.arrow:
          self.arrow.append(val)
      return self.arrow
    except Exception as e:
      print(e)


if __name__ == "__main__":
  my_data = Duplicate([1,2,2,2,1,3,3,3,4,4])
  result = my_data.remove_duplicate()
  print(result)







```

**@ second largest number**

```python
class SecondLargest:
  def __init__(self,data):
    self.data = data
    self.max1 = float( "-inf")
    self.max2 = float( "-inf")

  def second_largest(self):
    try:
      if self.data == "":
        raise ValueError("Error")
      else:
        for val in self.data:
          if val > self.max1:
            self.max2 = self.max1
            self.max1 = val
          elif val > self.max2 and val != self.max1:
            self.max2 = val

        return self.max2 , self.max1
    except Exception as e:
      print(e)


if __name__ == "__main__":
  my_data = SecondLargest([10,20,30,22,44,51,15])
  result = my_data.second_largest()
  print(result)
```
